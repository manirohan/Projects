```python
import sympy as sp

x = sp.symbols('x')
sigmoid = 1 / (1 + sp.exp(-x))

sp.plot(sigmoid, (x, -5, 5), title="Sigmoid Function", ylabel="f(x)", xlabel="x")
```

```python
x = sp.symbols('x')
sigmoid = sp.exp((-x**2))/(sp.pi)**(0.5)

sp.plot(sigmoid, (x, -3, 3), title="Function", ylabel="f(x)", xlabel="x")
```
